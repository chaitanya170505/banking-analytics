```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('./banking_data.csv')
df.head()
```

```python
df.shape
df.info()
```

```python
df.describe()
```

```python
df['Income Band'] = pd.cut(
    df['Estimated Income'],
    bins=[0, 100000, 300000, float('inf')],
    labels=['Low', 'Mid', 'High'],
    include_lowest=True
)

print(df['Income Band'].value_counts())
```

```python
categorical_cols = [
    'BR',
    'IA',
    'Gender',
    'Amount of Credit Cards',
    'Nationality',
    'Occupation',
    'Fee Structure',
    'Loyalty Classification',
    'Properties Owned',
    'Risk Weighting'
]


for col in categorical_cols:
    if col in df.columns:
        print("Value counts for", col)

        counts2 = df[col].value_counts(dropna=False)

        plt.figure(figsize=(6, 4))
        plt.bar(counts2.index.astype(str), counts2.values)
        plt.title(f"{col} Distribution")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.show()

```

```python
numerical_columns = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Loop through each column
for col in numerical_columns:
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.xlabel(col)         # Label x-axis with column name
    plt.ylabel('Frequency') # Label y-axis
    plt.title(f'Histogram of {col}')  # Title
    plt.show()
```

```python
# Select only numerical columns
numerical_columns = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]

# Compute correlation matrix
corr = df[numerical_columns].corr()

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# ðŸ“Š EDA Insights

## 1. High Activity Equals High Risk
- The bank's "risk" rating measures **financial activity and complexity**, not likelihood of failure.  
- A **higher risk score** usually indicates a **high-income customer** with multiple accounts and large balances.  
- As clients earn more, their **risk level rises naturally**, reflecting more engagement with the bank and management of complex financial tools.

## 2. The "Full-Wallet" Effect
- Successful customers **both save and borrow heavily**â€”they are not just choosing one strategy.  
- Clients with the **largest savings and deposits** also have **the largest loans**.  
- This suggests they leverage existing wealth to **borrow for business or investment purposes**, rather than out of cash shortage.

## 3. Success at Every Age
- **Age does not limit earnings or borrowing.**  
- High incomes and large loans are found **across all age groups**, from young adults to retirees.  
- The bankâ€™s clients are **high performers at every life stage**, with young professionals earning early and older clients maintaining wealth through smart investments.
